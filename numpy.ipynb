20.2 NumPy vícerozměrné pole

```python
brute_force = np.array([
    [0.2, 1.4412, 1.94724],   # čas
    [120, 240, 360],          # paměť (např. MB)
    [6, 8, 10]                # počet sedadel
])

monte_carlo = np.array([
    [0.22, 0.21, 0.22],
    [30, 30, 30],
    [20, 50, 100]
])

heuristika = np.array([
    [0.2, 0.342, 0.44],
    [40, 55, 70],
    [20, 40, 60]
])
```

```python
def vypis_casovou_slozitost(nazev, data):
    casova_slozitost = data[0]   # první řádek = čas
    print(f"{nazev}, časová složitost [sec]: {casova_slozitost}")
```

```python
vypis_casovou_slozitost("Brute force", brute_force)
vypis_casovou_slozitost("Monte carlo", monte_carlo)
vypis_casovou_slozitost("Heuristika", heuristika)
```


20.3 NumPy indexace

```python
print(brute_force[1, 1:-1])
print(monte_carlo[0, -5:])
```
