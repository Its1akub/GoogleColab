20.2 NumPy vícerozměrné pole

```python
brute_force = np.array([
    [0.2, 1.4412, 1.94724],   # čas
    [120, 240, 360],          # paměť (např. MB)
    [6, 8, 10]                # počet sedadel
])

monte_carlo = np.array([
    [0.22, 0.21, 0.22],
    [30, 30, 30],
    [20, 50, 100]
])

heuristika = np.array([
    [0.2, 0.342, 0.44],
    [40, 55, 70],
    [20, 40, 60]
])
```

```python
def vypis_casovou_slozitost(nazev, data):
    casova_slozitost = data[0]   # první řádek = čas
    print(f"{nazev}, časová složitost [sec]: {casova_slozitost}")
```

```python
vypis_casovou_slozitost("Brute force", brute_force)
vypis_casovou_slozitost("Monte carlo", monte_carlo)
vypis_casovou_slozitost("Heuristika", heuristika)
```


20.4 NumPy shaping and operations

```python
brute_force = np.array([1, 2, 3, 0.1932, 1.1247, 2.2435, 2243, 2124, 2143])
print(brute_force.shape)

brute_force_reshaped = brute_force.reshape(3, 3)
print(brute_force_reshaped)
print(brute_force_reshaped.shape)
```

```python
velikost, spotreba_cas, spotreba_pamet = np.array_split(brute_force, 3)

print(velikost)
print(spotreba_cas)
print(spotreba_pamet)
```

```python
spojene = np.concatenate((velikost, spotreba_cas, spotreba_pamet))
print(spojene)
```

```python
vysledek = np.stack((velikost, spotreba_cas, spotreba_pamet), axis=1)
print(vysledek)
print(vysledek.shape)
```

```python
vysledek = np.vstack((velikost, spotreba_cas, spotreba_pamet))
print(vysledek)
print(vysledek.shape)
```

```python
vysledek = np.hstack((velikost, spotreba_cas, spotreba_pamet))
print(vysledek)
print(vysledek.shape)
```

```python
vysledek = np.dstack((velikost, spotreba_cas, spotreba_pamet))
print(vysledek)
print(vysledek.shape)
```


20.6 NumPy searching

```python
data = np.array([20, 23, -100, -5, 30, 70, -18, 99, 81, 16,
                 45, 90, -39, -82, 75, 0, 16, 91, 48, 0, 70])
```

```python
idx_zaporna = np.where(data < 0)
print(idx_zaporna)
print(data[idx_zaporna])
```

```python
idx_rozsah = np.where((data >= -10) & (data <= -1))
print(data[idx_rozsah])
```

```python
idx_kombinace = np.where(((data >= 1) & (data <= 50)) | (data < 0))
print(data[idx_kombinace])
```

```python
from PIL import Image
```

```python
img = Image.open("8a68f104fc2394e6cca08922bbecf6d2.jpg")
```

```python
matrix = np.asarray(img)
```

```python
print("Typ pole:", type(matrix))
print("Datový typ matice:", matrix.dtype)
print("Tvar (shape) matice:", matrix.shape)
```

```python
print(matrix[:2, :5, :])  # ukáže prvních 2 řádky a 5 sloupců všech kanálů
```

```python
# Zmenšení 6× – vybereme každý šestý řádek a sloupec
mini_matrix = matrix[::6, ::6, :]
```

```python
Image.fromarray(mini_matrix.astype(np.uint8), 'RGB')
```

```python
# vytvoříme zapisovatelnou kopii mini_matrix
mini_matrix = mini_matrix.copy()

# nyní už lze upravit červenou složku
mini_matrix[:, :, 0] = 0  # nebo np.zeros([mini_matrix.shape[0], mini_matrix.shape[1]])

```

```python
Image.fromarray(mini_matrix.astype(np.uint8), 'RGB')

```
